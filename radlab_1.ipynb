```python
# Plot the result from radiation lab

# Import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
dose_data = np.array([[570, 583, 576.8, np.nan, np.nan, np.nan],
                      [585.8, 592, 585, 570, 536.7, 494],
                      [586.1, 592.3, 585.5, 567.8, 535, 496],
                      [575.6, 584.9, 578.7, 560, 529.9, 490.6]])

grid_number = np.array([[19, 18, 20, np.nan, np.nan, np.nan],
                      [16, 13, 10, 6, 2, 7],
                      [17, 14, 11, 3, 1, 4],
                      [21, 15, 12, 8, 5, 9]])
```

```python
fig, ax = plt.subplots(1,1,figsize = (9,5))
neg = ax.imshow(dose_data, cmap='cool', aspect = 'equal')
cbar = fig.colorbar(neg, location = 'right', shrink = 0.95)
cbar.set_label('Dose rate in water (mGy/s)', size = 14)
ax.set_axis_off()

for (j,i), label in np.ndenumerate(dose_data):
    if not(np.isnan(label)):
        ax.text(i,j, label, ha = 'center', va = 'center', fontsize = 14)
        
for (j,i), label in np.ndenumerate(grid_number):
    if not(np.isnan(label)):
        ax.text(i-0.45,j-0.3, int(label))

plt.show()
```
