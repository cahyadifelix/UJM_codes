```python
# Import libraries

import matplotlib.pyplot as plt
import numpy as np
```

```python
def extract(fname):
    return np.array([np.loadtxt(fname, skiprows=90, max_rows=500), np.loadtxt(fname, skiprows=1184)])
```

```python
otdr_pc50 = extract("2_2_51PC.asc")
otdr_pc50.shape
```

```python
def plot_otdr(fname1, fname2):
    data1 = extract(fname1)
    data2 = extract(fname2)
    
    fig, ax = plt.subplots(2,1, figsize = (8,5))
    ax[0].plot(data1[0,:,0], data1[0,:,1], label = "850 nm")
    ax[0].plot(data1[1,:,0], data1[1,:,1], label = "1300 nm", color = 'red')
    ax[0].grid()
    ax[0].set_xlabel("Position (km)", fontsize = 14)
    ax[0].set_ylabel("Reflected power (dB)", fontsize = 14)
    ax[0].set_xlim(0, 2.4)
    ax[0].set_ylim(0,25)
    ax[0].set_title("FC/PC connector")
    ax[0].legend()
    
    
    ax[1].plot(data2[0,:,0], data2[0,:,1], label = "850 nm")
    ax[1].plot(data2[1,:,0], data2[1,:,1], label = "1300 nm", color = 'red')
    ax[1].grid()
    ax[1].set_xlabel("Position (km)", fontsize = 14)
    ax[1].set_ylabel("Reflected power (dB)", fontsize = 14)
    ax[1].set_xlim(0, 2.4)
    ax[1].set_ylim(0,25)
    ax[1].set_title("FC/APC connector")
    ax[1].legend()
    
    fig.tight_layout(pad = 0.5)
    
    
```

```python
plot_otdr("2_2_51PC.asc", "2_2_51APC.asc")
```

```python
plot_otdr("2_2_51PC_625.asc","2_2_51APC_625.asc")
```
