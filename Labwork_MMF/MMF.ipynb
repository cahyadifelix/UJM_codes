```python
# Import libraries

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
radi_arr = np.array([5,10,15,20,25])
radi_power = np.array([625, 633, 640, 640, 640])
```

```python
fig_radi, ax_radi = plt.subplots(1,1, figsize = (8,5))
ax_radi.plot(radi_arr, radi_power, color = 'black', linestyle = '--', marker = 'o', markerfacecolor = 'red')
ax_radi.set_xlabel("Curvature radius (mm)", fontsize = 14)
ax_radi.set_ylabel("Power ($\mu$W)", fontsize = 14)
ax_radi.grid()
```

```python
trans_pwr = np.array([110, 103, 67, 26, 9, 2.1, 0.9, 0.6, 0.39, 0.16, 0.033])

trans_pwr_dB = -10*np.log10(trans_pwr/trans_pwr[0])
trans_delta = np.array([0,20,40,60,80,100,120,140,160,180, 200])
```

```python
long_pwr = np.array([160, 114, 91, 68, 49, 34, 29, 26, 16.5, 15.2, 11.8])

long_pwr_dB = -10*np.log10(long_pwr/long_pwr[0])
long_delta = trans_delta*10
```

```python
fig_coupling, ax_coupling = plt.subplots(2,1, figsize = (8,8))
ax_coupling[0].plot(trans_delta, trans_pwr_dB, color = 'black', linestyle = '--', marker = 'o', markerfacecolor = 'red')
ax_coupling[0].grid()
ax_coupling[0].set_xlabel("Transversal misalignment ($\mu$m)", fontsize = 14)
ax_coupling[0].set_ylabel("Excess loss (dB)", fontsize = 14)
ax_coupling[0].set_xlim([0,200])
#ax_coupling[0].set_yscale('log')

ax_coupling[1].plot(long_delta, long_pwr_dB, color = 'black', linestyle = '--', marker = 'o', markerfacecolor = 'red')
ax_coupling[1].grid()
ax_coupling[1].set_xlabel("Longitudinal misalignment ($\mu$m)", fontsize = 14)
ax_coupling[1].set_ylabel("Excess loss (dB)", fontsize = 14)
ax_coupling[1].set_xlim([0,2000])

```

```python
pwr_cb = np.array([-4.64, -4.54, -4.52, -4.48])
length_removed = np.array([0, 1.41, 2.34, 3.54])

pwr_cb_rel = pwr_cb-pwr_cb[0]

# Fitting
res = linregress(length_removed, pwr_cb_rel)
```

```python
fig_cb, ax_cb = plt.subplots(1,1,figsize = (8,5))
ax_cb.scatter(length_removed, pwr_cb_rel, color = 'red', edgecolors='black', label = "Measured")
ax_cb.axline((0,res.intercept), slope = res.slope, color = 'black', linestyle = '--', label = f"Fit: {res.intercept*1000:.2f} dB/km")
ax_cb.grid()
ax_cb.set_xlabel("Removed length (m)", fontsize = 14)
ax_cb.set_ylabel("Relative power (dB)", fontsize = 14)
ax_cb.legend()
```
