```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import hilbert
import matplotlib
```

```python
def cos_wave(amplitude, phase, k, n_dots = 1000):
    return amplitude*np.cos(k*np.linspace(-np.pi, np.pi, n_dots)+phase)
```

```python
n_dots = 100000

x_array = np.linspace(-np.pi, np.pi, n_dots)

start_k = 80
k_span = 100

k_array = np.arange(start_k, start_k+k_span,step = 2)
print(k_array)
```

```python
#%matplotlib qt

fig_test, ax_test = plt.subplots(2,1, figsize = (8,6))
total_wave = np.zeros(n_dots)
shift_array = []
for i, k in enumerate(k_array):
    shift = 0
    #shift = -((i-(k_span/4))**3)*1e-4
    shift = k*(i-(k_span/4))*2.4e-3
    shift_array.append((-1/k)*shift)
    wave = cos_wave(1,shift, k, n_dots)
    ax_test[0].plot(x_array, wave+4*i, color = 'black')
    total_wave = total_wave + wave
    
ax_test[0].set_xticks([-np.pi, -np.pi/2, 0 ,np.pi/2, np.pi])
ax_test[0].set_xticklabels(['$-\pi$', '$-\pi/2$', 0, '$\pi/2$', '$\pi$'])
ax_test[0].set_xlim((-np.pi/2, np.pi/2))
ax_test[0].tick_params(left = False, labelleft = False)
ax_test[0].plot(shift_array, np.arange(0,len(k_array))*4+1, color = 'red')
#ax_test[0].plot(shift_array, increment_array*3)
#ax_test[0].axvline(0, color = 'black', alpha = 0.5)
#ax_test[0].grid()
#print(np.power(np.linspace(1,i-1),3)*1e-4)


ax_test[1].plot(x_array,total_wave, color = 'black')
ax_test[1].set_xticks([-np.pi, -np.pi/2, 0 ,np.pi/2, np.pi])
ax_test[1].set_xticklabels(['$-\pi$', '$-\pi/2$', 0, '$\pi/2$', '$\pi$'])
ax_test[1].set_xlim((-np.pi/2, np.pi/2))
ax_test[1].set_ylim(-60,60)
ax_test[1].tick_params(left = False, labelleft = False)
ax_test[1].axvline(0, color = 'black', alpha = 0.5)
#ax_test[1].grid()ss
```

```python
fig, ax = plt.subplots(1,1, figsize = (8,6))
for i in range(2):
    ax.plot(x_array, cos_wave(1,i*2*np.pi, 10, n_dots), color = 'black')
```
